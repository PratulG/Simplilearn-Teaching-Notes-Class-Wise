#### Encapsulation

```python
#Protected Members
class Base:
    def __init__(self):
        
        #Protected Members
        self._a = 2
        
# Creating a derived class
class Derived(Base):
    def __init__(self):
        
        #Calling the base class
        Base.__init__(self)
        print('Calling protected member of base class: ',self._a)
        
        # Modify the protected variable
        self._a = 3
        print('Calling modified protected member outside class:', self._a)

obj1 = Derived()

obj2 = Base()

#Calling the protected members
print("Accessing protected members of obj1: ",obj1._a)

# Accessing the protected member variable outside of the class
print("Accessing protected members of obj2: ",obj2._a)
```

```python
# Private Method
class Base:
    def __init__(self):
        self.a = 'Simplilearn'
        self.__c = 'Simplilearn'
        
# Creating a derived class
class Derived(Base):    
    def __init__(self):
        
        Base.__init__(self)
        print('Calling private member of base class: ')
        print(self.__c)
        
# Calling or Driver Code
obj1 = Base()
print(obj1.a)
```

```python
# Private Method
class Base:
    def __init__(self):
        self.a = 'Simplilearn'
        self.__c = 'Simplilearn'
        
# Creating a derived class
class Derived(Base):    
    def __init__(self):
        
        Base.__init__(self)
        print('Calling private member of base class: ')
        print(self.__c)
        
# Calling or Driver Code
obj1 = Base()
print(obj1.a)

#Uncommenting this will raise an error
#obj2 = Derived() : This will raise an AttributeError as private member of base class is called inside derived class
#print(obj2.a)
```
