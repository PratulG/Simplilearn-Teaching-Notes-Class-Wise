```python
from statsmodels.stats.weightstats import ttest_ind
import numpy as np
import pandas as pd
import pingouin as pg
```

```python
data_group1 = np.array([160,150,160,156.12,163.24,160.56,168.56,174.12,167,165])
data_group2 = np.array([164,153,168,166.12,173.24,150.56,158.56,171,157,170])
```

```python
data_group1
```

```python
# Conductimg two sample ttest
result = pg.ttest(data_group1,data_group2,correction = True)
result
```

Group 1 and Group 2 heights are statistically equivalent

# Two way Anova

```python
pip install statsmodels
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd
import numpy as np
```

```python
dataframe = pd.DataFrame({'Fertilizers':np.repeat(['daily','weekly'],15),'Watering': np.repeat(['daily','weekly'],15),
                          'Height':[14,16,15,15,16,13,12,11,
                                    14,15,16,16,17,18,14,13,
                                    14,14,14,15,16,16,17,18,
                                    14,13,14,14,14,15]})
```

```python
dataframe
```

```python
# Performing two way Anova
model = ols('Height ~ C(Fertilizers) + C(Watering)',
            data = dataframe).fit()
result = sm.stats.anova_lm(model,type = 2)

result
```


*   p-value for Fertilizers is 0.913305

*   p-value for Watering is 0.990865


*   Statistically Fertilizers and Watering have least significance on Height of plant which may infer there can be other factors like Weather, Geography , Temperature which can affect plant height




# Dimensionality Reduction Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Wine.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,:-1].values
X
```

```python
y = dataset[['Customer_Segment']].values
y
```

```python
dataset.isna().sum()
```

```python
dataset.info()
```

```python
dataset.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 7)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

## Applying PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
X_train
```

```python
# Model Fit
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 7)
classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
