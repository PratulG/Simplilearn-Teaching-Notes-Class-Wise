## While Loop

```python
x = 0

while x < 10:
  print('x is currently: ',x)
  print('x is still less than 10, adding 1 to x')
  x = x + 1 # x+= 1
```

```python
x = 0

while x < 10:
  print('x is currently: ',x)
  print('x is still less than 10, adding 1 to x')
  x = x + 1 # x+= 1

else:
  print('stop')
```

```python
x = 0

while x < 10:
  print('x is currently: ',x)
  print('x is still less than 10, adding 1 to x')
  x = x + 1 # x+= 1
  if x == 3:
    break
```

```python
x = 0

while x < 10:
  print('x is currently: ',x)
  print('x is still less than 10, adding 1 to x')
  x+= 1 # x = x + 1
  if x == 3:
    print('x==3')
  elif x == 5:
    print('x==5')
  elif x == 7:
    print('x==7')      
  else:
    print('continue')
    continue
```

## Useful Operator

```python
index_count = 0

for letter in 'abcde':
  print("At index {} the letter is {}".format(index_count,letter))
  index_count += 1
  
```

```python
list(enumerate('abcde'))
```

```python
for i,letter in enumerate('abcde'):
    print("At index {} the letter is {}".format(i,letter))
```
