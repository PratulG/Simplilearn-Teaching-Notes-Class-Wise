## Merge

```python
# Case 1 :Merging with one condition
data1 = {'key':['K0','K1','K2','K3'],'Name':['Jai','Princi','Gaurav','Anuj'],'Age':[27,24,22,32]}
data2= {'key':['K0','K1','K2','K3'],'Address':['Nagpur','Kanpur','Allahabad','Kannauj'],
        'Qualification':['Btech','BA','BCom','B.hons']}
```

```python
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)
```

```python
df
```

```python
df1
```

```python
pd.merge(df,df1,on = 'key')
```

```python
# Case 2 : Memrging with multiple condition
data1 = {'key':['K0','K1','K2','K3'],
         'key1':['K0','K1','K0','K1'],'Name':['Jai','Princi','Gaurav','Anuj'],'Age':[27,24,22,32]}
data2= {'key':['K0','K1','K2','K3'],
        'key1':['K0','K0','K0','K0'],'Address':['Nagpur','Kanpur','Allahabad','Kannauj'],
        'Qualification':['Btech','BA','BCom','B.hons']}
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)
df
```

```python
df1
```

```python
pd.merge(df,df1,on = ['key','key1'])
```

```python
# Case 3 :left
data1 = {'key':['K0','K1','K2','K3'],
         'key1':['K0','K1','K0','K1'],'Name':['Jai','Princi','Gaurav','Anuj'],'Age':[27,24,22,32]}
data2= {'key':['K0','K1','K2','K3'],
        'key1':['K0','K0','K0','K0'],'Address':['Nagpur','Kanpur','Allahabad','Kannauj'],
        'Qualification':['Btech','BA','BCom','B.hons']}
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)
df
```

```python
df1
```

```python
pd.merge(df,df1,how = 'left',on = ['key','key1'])
```

```python
pd.merge(df,df1,how = 'right',on = ['key','key1'])
```

```python
df1
```

```python
pd.merge(df,df1,how = 'outer',on = ['key','key1'])
```

```python
pd.merge(df,df1,how = 'inner',on = ['key','key1'])
```

```python
pd.merge(df,df1,how = 'left',on = ['key'])[['Age','Qualification']]
```
